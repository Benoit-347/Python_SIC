```python
print(""" This is David's fruit shop.
 1: Apple( price : 5000 won)
 2: Grape( price : 6000 won)
 3: Melon( price : 8000 won)
 4: Orange( price : 2000 won)""")
dict_fruit_price = {"apple": 5000, "grape": 6000, "melon": 8000, "orange": 2000}

while True:
    user_input = input("Enter fruit name: ")
    if user_input in dict_fruit_price:
        break
    else:
        print("Invalid fruit name entered. try again-")
price = dict_fruit_price[user_input]

print(f"Enter quantity: ")
quantity = int(input())

total_price = price * quantity
print(f"Total price to pay: {total_price}")

print("Enter cash you give: ", end = '')
Money_given = int(input())
print(Money_given)
while (Money_given < total_price):
    print("Insufficent money, please give more: ", end = '')
    Money_given += int(input())
    print(Money_given)

remaining_change = Money_given - total_price
print(f"Here you go your change: {remaining_change}")
```

```python
import random

for _ in range(10):
    #sim user input
    #get operator
    operator_index = random.randint(0, 3)
    list_operators = ['+', '-', '*', '/']
    operator = list_operators[operator_index]

    # get operands
    sign = random.randint(0, 1)
    if sign == 0:
        sign = 1
    else:
        sign = -1

    operand_1 = random.randint(0, 100)*sign
    operand_2 = random.randint(1, 100)

    match(operator):
        case '+':
            result = operand_1 + operand_2
        case '-':
            result = operand_1 - operand_2
        case '*':
            result = operand_1 * operand_2
        case '/':
            result = operand_1 / operand_2

    print(f"Question: {operand_1} {operator} {operand_2}")
    print(f"Answer: {result}")
```

```python
substat_names = ["atk", "def", "hp", "crit rate", "crit dmg", "elemental mastery", "energy recharge"]
stat_combos = []

for stat_1 in substat_names:
    for stat_2 in substat_names:
        for stat_3 in substat_names:
            for stat_4 in substat_names:
                stat_combos.append([stat_1, stat_2, stat_3, stat_4])

# literally fastest method using py VMM (cpython)
import itertools
combos = itertools.product(substat_names, substat_names, substat_names, substat_names)
combos = list(combos)
```
